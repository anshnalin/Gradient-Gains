```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "House Area": [6274,6689,4276,5130,6035,2946,4727,6873,3911,5689,5459,4508,3589,7212,7454,3366],
    "Rooms":      [8,5,8,9,6,1,10,7,4,5,5,4,8,3,9,6],
    "Price":      [2100,2200,1400,1700,2000,1000,1600,2300,1300,1900,1800,1500,1200,2400,2500,1100]
}

df = pd.DataFrame(data)
X = df[["House Area", "Rooms"]].values
y = df["Price"].values.reshape(-1, 1)

m = X.shape[0]
X_bias = np.hstack((np.ones((m, 1)), X))

theta = np.linalg.inv(X_bias.T @ X_bias) @ X_bias.T @ y
y_pred=X_bias@theta
comparison = pd.DataFrame({
    "Actual Price": y.flatten(),
    "Predicted Price": y_pred.flatten()
})

plt.figure()
plt.scatter(y, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.grid(True)
plt.show()

plt.figure()
plt.scatter(df["House Area"], y, label="Actual")
plt.scatter(df["House Area"], y_pred, label="Predicted")
plt.xlabel("House Area")
plt.ylabel("Price")
plt.title("House Area vs Price")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.scatter(df["Rooms"], y, label="Actual")
plt.scatter(df["Rooms"], y_pred, label="Predicted")
plt.xlabel("Number of Rooms")
plt.ylabel("Price")
plt.title("Rooms vs Price")
plt.legend()
plt.grid(True)
plt.show()

new_house = np.array([[1, 10000, 10]]) #just an example, i think i gotta remove this.
predicted_price = new_house @ theta
predicted_price

```
